### Univariate analysis

```python
df_setosa=df.loc[df[' flower type']=='Iris-setosa']
# give a space before flower type

df_virginica=df.loc[df[' flower type']=='Iris-virginica']
df_versicolor=df.loc[df[' flower type']=='Iris-versicolor']
```

```python
plt.plot(df_setosa['sepal length'], np.zeros_like(df_setosa['sepal length']),'o')
plt.plot(df_virginica['sepal length'], np.zeros_like(df_virginica['sepal length']),'*')
plt.plot(df_versicolor['sepal length'], np.zeros_like(df_versicolor['sepal length']),'+')
plt.xlabel("petal length")
plt.show()
```

### Bivariate analysis

```python
sns.FacetGrid(df,hue=" flower type", size=5).map(plt.scatter, "sepal length", "sepal width").add_legend();
plt.show()
```

```python
# give space to both ends of the petal length
sns.FacetGrid(df,hue=" flower type", size=5).map(plt.scatter, ' petal length ', "sepal width").add_legend();
plt.show()
```

### Multivariate analysis

```python
sns.pairplot(df, hue= ' flower type', size=3)
```
